```python
import os
import pandas as pd

def load_and_merge_logs(log_dir):
    all_logs = []
    for file in os.listdir(log_dir):
        if file.endswith('.csv'):
            file_path = os.path.join(log_dir, file)
            user_id = file.split('_')[1]  # Extract user ID from the file name
            user_id = user_id.split('.')[0]  # Extract user ID from the file name
        
            log_data = pd.read_csv(file_path, sep=';', on_bad_lines='skip')
            
            log_data['user_id'] = user_id
            
            log_data = log_data[~log_data['user_id'].isin(['14', '23'])]
            
            study_end_index = log_data[log_data[' Event Name'] == ' Study End'].index
            if not study_end_index.empty:
                log_data = log_data.iloc[:study_end_index[0] + 1]
            
            all_logs.append(log_data)
    
    merged_logs = pd.concat(all_logs, ignore_index=True)
    return merged_logs

log_data = load_and_merge_logs('LogFiles')
```

```python
# Function to clean and process pre-test data
def clean_pre_test_data(file_path):
    data = pd.read_csv(file_path)
    # Drop unnecessary columns
    columns_to_drop = [
        col for col in data.columns if '[Score]' in col or '[Feedback]' in col or col in ['Timestamp', 'Total score']
    ]
    data = data.drop(columns=columns_to_drop, errors='ignore')
    # Rename columns for easier access
    data = data.rename(columns={
        'Please enter your Participant ID': 'participant_id',
        'Please select your age group (Please select one)': 'age_group',
        'Please select your gender (Please select one) ': 'gender',
        'What is your dominant hand': 'dominant_hand',
        'Are you experiencing any discomfort (such as headache, fatigue, or eye strain) that could affect your participation in this test? (Please select one)': 'discomfort',
        'Do you consent to participate in this study under the conditions provided in the consent form? (Please select one)': 'consent',
        'How many hours have you spent using VR in the past year? (Please select one) ': 'vr_experience',
        'Do you experience motion sickness, dizziness, or discomfort when using VR? (Please select one) ': 'motion_sickness',
        'On a scale of 1 - 7, how confident are you using new technology ( 1= not confident, 7 = very confident)': 'tech_confidence',
        'How familiar are you with aiming or shooting tasks on electronic displays? (Please select one)': 'aiming_familiarity',
        'Do you have any hearing or visual impairments or conditions that could affect your ability to see or interact with images in VR?': 'impairments'
    })
    return data
```

```python
results = []

for i, row in df.iterrows():
    if row[" Event Name"] == " Target spawned" or row[" Event Name"] == " Target respawned":
        spawn_time = row["Time"]
        accuracy = None
        feedback = None
        button = None
        response_time = None
        
        for j in range(i + 1, len(df)):
            next_row = df.iloc[j]    
            if next_row[" Event Name"] == " Button Pressed":
                accuracy = next_row[" Accuracy"]
                time_row = df.iloc[j+1]
                response_time = time_row["Time"] - spawn_time
                break

        for j in range(i + 1, len(df)):
            next_row = df.iloc[j]
            if next_row[" Event Name"] == " Button Pressed":
                feedback = next_row[" Feedback"]
                button = next_row[" Button"]
                break

        if response_time is not None and accuracy is not None:
            results.append({
                "Accuracy": accuracy,
                "Response Time": response_time,
                "Feedback Modality": feedback,
                "Button": button
            })

results_df = pd.DataFrame(results)
results_df = results_df[results_df["Feedback Modality"] != " Practice"].reset_index(drop=True)
#print(results_df[['Response Time', 'Feedback Modality']])
```

```python
import pandas as pd

def remove_first_occurrence_of_new_feedback(results_df):
    results_df = results_df.drop(index=0)
    previous_feedback = None
    rows_to_remove = []

    for index, row in results_df.iterrows():
        current_feedback = row['Feedback Modality']

        if previous_feedback is not None and current_feedback != previous_feedback:
            rows_to_remove.append(index)  # Die erste Zeile nach dem Wechsel entfernen

        previous_feedback = current_feedback

    cleaned_results_df = results_df.drop(rows_to_remove)

    return cleaned_results_df

results_df = remove_first_occurrence_of_new_feedback(results_df)

#print(results_df_cleaned[['Response Time', 'Feedback Modality']])
```

### 1. Hypothesis: Feedback modalities with combinations of feedback modalities will result in higher shooting accuracy and faster response times compared to single-modality feedback
##### 1.1: Feedback modalities with combinations of feedback modalities will result in higher shooting accuracy compared to single-modality feedback
##### 1.2: Feedback modalities with combinations of feedback modalities will result in faster response times compared to single-modality feedback
##### Mann-Whitney-U-Test since not normally distributed and we want to compare two groups

```python
from scipy.stats import kruskal
from scipy.stats import mannwhitneyu

# Separate single modalities and combinations
single_modalities = results_df[~results_df['Feedback Modality'].str.contains(r'\+')]
combinations = results_df[results_df['Feedback Modality'].str.contains(r'\+')]

stat, p = mannwhitneyu(single_modalities['Accuracy'], combinations['Accuracy'], alternative='two-sided')
print(f"Mann-Whitney-U-Test for Accuracy (Single vs Combinations): Teststatistik = {stat}, p-Wert = {p}")

stat, p = mannwhitneyu(single_modalities['Response Time'], combinations['Response Time'], alternative='two-sided')
print(f"Mann-Whitney-U-Test for Response Time (Single vs Combinations): Teststatistik = {stat}, p-Wert = {p}")

```

### 2. Hypothesis: Feedback modalities with haptic feedback show less accuracy

```python
from scipy.stats import mannwhitneyu

# Separate the data for haptic feedback and all other modalities
haptic = results_df[results_df['Feedback Modality'] == ' haptic']['Accuracy'].dropna()
non_haptic = results_df[results_df['Feedback Modality'] != 'haptic']['Accuracy'].dropna()

# Perform the Mann-Whitney-U Test
stat, p = mannwhitneyu(haptic, non_haptic, alternative='two-sided')
print(f"Mann-Whitney-U-Test for Accuracy (Haptic vs. Others): Test Statistic = {stat}, p-Value = {p}")
```

### 3. Hypothesis: Same-hand trigger will result in faster response times

```python
# Separate the data for "Trigger" (same-hand) and "X" (other-hand)
trigger = results_df[results_df['Button'] == ' Trigger']['Response Time'].dropna()
other_hand = results_df[results_df['Button'] == ' X']['Response Time'].dropna()

stat, p = mannwhitneyu(trigger, other_hand, alternative='less')
print(f"Mann-Whitney-U-Test for Response Time (Trigger vs. X): Test Statistic = {stat}, p-Value = {p}")
```
